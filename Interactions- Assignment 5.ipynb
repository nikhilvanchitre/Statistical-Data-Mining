```r
data<-read.csv(choose.files())
```

```r
data
```

```r
m1<-lm(data$AmountSpent~data$Catalogs*data$Salary)
summary(m1)
```

```r
m2<-lm(data$AmountSpent~data$Catalogs+data$Salary)
summary(m2)
```

# Answers for Q1
1) m1 is better than m2 as adjusted R-squared is more for m1.
2) Yes, Salary and Catalogs are both significant when interaction is not there, that is in model 1. If interaction term is considered, then catalogs is not significant.
3) m1: Amount spent will be decreased but not significantly. 
   m2: amount spent will increase by 51.7.

```r
m3<-lm(data$AmountSpent~data$Salary*as.factor(data$Location)+data$Catalogs*as.factor(data$Location)+data$Children*as.factor(data$Location))
summary(m3)
```

# Answers for Q2
Regression Equation:


a) Amount Spent = -310.3 + 0.01737 X Salary + 39.98 X Catalogs - 191.7 X Children -0.02 X Far - 0.01212 X Salary X Far + 5.89 X Far X Catalogs - 0.049 X Far X Children


b) Salary and Location interaction term are significant.

```r
m4<-lm(data$AmountSpent~data$Salary*data$Catalogs+data$Catalogs*data$Children+data$Catalogs*as.factor(data$Location))
summary(m4)
```

# Answers for Q4

a) m4 is slightly better than m3 as adjusted R-squared value is slightly greater for m4.

b) all three interaction terms are significant as all the three p-values are less than 0.05.
